```python
import pandas as pd
import numpy as np
```

```python
input_1=pd.read_csv('https://docs.google.com/spreadsheets/d/1CD_ZszagDlCrYEvKYuav_CxnARYHgVVODdgeHyvzFD4/export?gid=1473172222&format=csv')
input_2=pd.read_csv('https://docs.google.com/spreadsheets/d/1CD_ZszagDlCrYEvKYuav_CxnARYHgVVODdgeHyvzFD4/export?gid=99428384&format=csv')
```

```python
input_1.dropna(axis=1,how='all',inplace=True)
input_1.dropna(how='all',inplace=True)
input_2.dropna(axis=1,how='all',inplace=True)
input_2.dropna(how='all',inplace=True)
input_1.reset_index(drop=True,inplace=True)
input_2.reset_index(drop=True,inplace=True)
```

```python
inpu=input_1.iloc[8:8+21,:]
inpu.columns=list(input_1.iloc[7,:])
inpu=inpu.set_index('S No')
inpu['User ID'] = inpu['User ID'].astype('float32')
inpu['Team Name']=inpu['Team Name'].str.strip()
inpu['Name']=inpu['Name'].str.strip()
inpu
```

```python
inpu2=input_2.iloc[6:6+21,:5]
inpu2.columns=list(input_2.iloc[5,:5])
inpu2=inpu2.set_index('S No')
inpu2['uid'] = inpu2['uid'].astype('float32')
inpu2['total_statements'] = inpu2['total_statements'].astype('float32')
inpu2['total_reasons'] = inpu2['total_reasons'].astype('float32')
inpu2
```

```python
df=inpu.join(inpu2.iloc[:,2:])
df
```

```python
output_1=df.groupby(df['Team Name'].str.lower())[['total_statements','total_reasons']].mean().round(2).sort_values(['total_statements','total_reasons'],ascending=[False,False]).reset_index()
output_1.index=[i for i in range(1,len(output_1)+1)]
output_1.index.name='Team Rank'
output_1.columns=['Thinking Teams Leaderboard','Average Statements','Average Reasons']
output_1
```

```python
output_2=inpu2.sort_values(['total_statements','total_reasons'],ascending=[False,False]).reset_index()
output_2.index=[i for i in range(1,len(output_2)+1)]
output_2.index.name='Rank'
output_2.drop(columns='S No',inplace=True)
output_2.columns=['Name','UID','No. of Statements','No. of Reasons']
output_2
```
